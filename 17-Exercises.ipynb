### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
# Exercise 1

import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
# company_sales.describe()
company_sales
````

````python
#Exercise 1 for homework
fig, ax1 = plt.subplots()

ax1.set_title('Total profit of All Months')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Total Profit')

x = company_sales['month_number'].values
y = company_sales['total_profit'].values

plt.plot(x,y)
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
fig, ax1 = plt.subplots()

ax1.set_title('Total Profit of All Months')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Sold units number')

x = company_sales['month_number'].values
y = company_sales['total_units'].values


for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(x, y, ':p', color='red',
         markersize=5, linewidth=3,
         markerfacecolor='red',
          markeredgecolor='red',
         markeredgewidth=5)
plt.legend(['Total Profit Over Time'], loc='lower right')
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3
fig, ax1 = plt.subplots()

ax1.set_title('All Product Sales')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Sold units number')

x = company_sales['month_number'].values
y1 = company_sales['facecream'].values
y2 = company_sales['facewash'].values
y3 = company_sales['toothpaste'].values
y4 = company_sales['bathingsoap'].values
y5 = company_sales['shampoo'].values
y6 = company_sales['moisturizer'].values

plt.plot(x,y1, label = 'facecream')
plt.plot(x,y2, label='facewash')
plt.plot(x,y3, label='toothpaste')
plt.plot(x,y4, label='bathingsoap')
plt.plot(x,y5, label='shampoo')
plt.plot(x,y6, label ='moisturizer')
plt.legend(loc='upper left');
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
# Exercise 5
ax = plt.axes(label="5")
plt.bar(company_sales.month_number, company_sales.facewash, color='blue', label='Facewash', width = 0.25)
plt.bar(company_sales.month_number + .25, company_sales.facecream, color='yellow', label='Facecream', width = 0.25)
plt.legend(loc='upper left')
ax.set(xlabel='Month Number', ylabel='Number of Units Sold',
       title='Exercise 5');
````
